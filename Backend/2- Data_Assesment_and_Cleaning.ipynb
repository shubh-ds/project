#### Column assesment: Possession Status

```python
# Format values
df['Possession Status'] = df['Possession Status'].str.strip()

# Count unique values
df['Possession Status'].value_counts()
```

```python
# Format
df['Possession Status'] = df['Possession Status'].str.replace(" '", " ")
```

```python
# Replace past dates with 'Ready to move' status
df['Possession Status'].replace({'Immediately': 'Ready to Move', 
                                 'Dec 24': 'Ready to Move', 
                                 'Dec 23': 'Ready to Move', 
                                 'Jun 24': 'Ready to Move', 
                                 'Mar 25': 'Ready to Move',
                                 'Mar 24': 'Ready to Move',
                                 'Sep 23': 'Ready to Move',
                                 'Jun 25': 'Ready to Move',
                                 'Apr 24': 'Ready to Move',
                                 'Nov 23': 'Ready to Move'
                                 }, inplace=True)
```

```python
# Updated value counts
df['Possession Status'].value_counts()
```

```python
# Drop row having 'Select Date' Possesion status
index_to_drop = df[df['Possession Status'] == 'Select Date'].index
df.drop(index=index_to_drop, inplace=True)

# Check updated shape
df.shape
```

#### Column assesment: Availability Starts From

```python
# Format values
df['Availability Starts From'] = df['Availability Starts From'].str.strip()
```

```python
# 'Ready to Move' properties indexes
ready_to_move_indexes = df[df['Possession Status'] == 'Ready to Move'].index

# Set 'Availability Starts From' to 'Immediately' where Possesion status is 'Ready to Move'
df.loc[ready_to_move_indexes, 'Availability Starts From'] = 'Immediately'
```

```python
# Count unique values
df['Availability Starts From'].value_counts()
```

```python
# Format values
df['Availability Starts From'] = df['Availability Starts From'].str.replace(" '", " ")

# Update past dates to 'Immediately'
df['Availability Starts From'].replace({ 
                                 'Dec 22': 'Immediately',

                                 'Mar 23': 'Immediately',
                                 'Apr 23': 'Immediately',
                                 'May 23': 'Immediately',
                                 'Jun 23': 'Immediately',
                                 'Jul 23': 'Immediately',
                                 'Aug 23': 'Immediately',
                                 'Sep 23': 'Immediately',
                                 'Oct 23': 'Immediately',
                                 'Nov 23': 'Immediately',
                                 'Dec 23': 'Immediately',

                                 'Jan 24': 'Immediately',
                                 'Feb 24': 'Immediately',
                                 'Mar 24': 'Immediately',
                                 'Apr 24': 'Immediately',
                                 'May 24': 'Immediately',
                                 'Jun 24': 'Immediately',
                                 'Jul 24': 'Immediately',
                                 'Aug 24': 'Immediately',
                                 'Sep 24': 'Immediately',
                                 'Oct 24': 'Immediately',
                                 'Nov 24': 'Immediately',
                                 'Dec 24': 'Immediately',

                                 'Jan 25': 'Immediately',
                                 'Feb 25': 'Immediately',
                                 'Mar 25': 'Immediately',
                                 'Apr 25': 'Immediately',
                                 'May 25': 'Immediately',
                                 'Jun 25': 'Immediately',
                                 'Jul 25': 'Immediately'
                                 }, inplace=True)
```

```python
# Update 'Possesion Status' of properties to 'Ready to Move' where 'Availability Starts From' is 'Immediately'
immediate_availability_indexes = df[df['Availability Starts From'] == 'Immediately'].index
df.loc[immediate_availability_indexes, 'Possession Status'] = 'Ready to Move'
```

```python
# Get properties where date is provided for Possesion status
possesion_status_date_provided_indexes = df[~((df['Possession Status'] == 'Ready to Move') | (df['Possession Status'] == 'Under Construction') | (df['Possession Status'].isna()))].index

# Copy these dates to 'Availability Starts From'
df.loc[possesion_status_date_provided_indexes, 'Availability Starts From'] = df.loc[possesion_status_date_provided_indexes, 'Possession Status']
```

```python
# Replace future dates in 'Possesion Status' with 'Under Construction'
idx = df[(df['Possession Status'] != 'Ready to Move') & (df['Possession Status'] != 'Under Construction') & (df['Possession Status'].notna())].index
df.loc[idx, 'Possession Status'] = 'Under Construction'
```

```python
# Replace NaN values in 'Possession Status' and 'Availability Starts From' with 'Not Available, Contact Developer'
df['Possession Status'].fillna('Not Available, Contact Developer', inplace=True)
df['Availability Starts From'].fillna('Not Available, Contact Developer', inplace=True)
```
